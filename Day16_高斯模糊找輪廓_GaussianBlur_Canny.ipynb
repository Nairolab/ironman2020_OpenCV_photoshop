```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from IPython.display import clear_output
```

```python
def show_img(img):
    plt.style.use('dark_background')
#     plt.figure(figsize=(10,10)) 
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.show()
```

```python
# 高斯模糊(高斯平滑)，與計算核心大小
def do_GaussianBlur(gray):
    kernel_size = 5
    blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size), 0)
    return blur_gray
```

```python
# Canny邊緣運算
def do_Canny(blur_gray):
    low_threshold = 10
    high_threshold = 90
    edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
    return edges
```

```python
def img_processing(img):
    # do something here
    
    print("Gray image:")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # start from BGR -> gray
    show_img(gray)

    print("GaussianBlur image:")
    blur_img = do_GaussianBlur(gray)
    show_img(blur_img)
    
    print("Canny edges:")
    edges = do_Canny(blur_img)
    show_img(edges)
    
#     # ---------------- 印出結果圖表 ---------------- #

    plt.figure(figsize=(15,10)) 
    
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Origin picture", {'fontsize':20}) 
    
    plt.subplot(1, 3, 2)
    plt.imshow(cv2.cvtColor(blur_img, cv2.COLOR_BGR2RGB))
    plt.title("GaussianBlur", {'fontsize':20})  

    
    plt.subplot(1, 3, 3)
    plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
    plt.title("Canny", {'fontsize':20})
    
    plt.show()
    
```
