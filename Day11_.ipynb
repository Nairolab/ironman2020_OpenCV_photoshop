```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from IPython.display import clear_output
```

```python
def show_img(img):
#     plt.figure(figsize=(15,15)) 
    plt.figure(figsize=(10,10)) 
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.show()
```

```python
def draw_lines(img):
    cv2.line(img, (0, 0), (255, 255), (0, 0, 255), 5)
    
    
```

```python
def draw_lines(img):
    cv2.line(img, (0, 0), (255, 255), (0, 0, 255), 5)
    
    
```

```python
def draw_points(img):
    
    
    point_size = 10
    point_color = (0, 0, 255) # BGR
    thickness = 4 # 可以为 0 、4、8
    
#     img.fill(200)

    
    cv2.rectangle(img, (20, 60), (120, 160), (0, 255, 0), 2)
    cv2.rectangle(img, (40, 80), (100, 140), (0, 255, 0), -1)

    # 黃色圓圈，線條寬度為 3 px
    cv2.circle(img,(90, 210), 30, (0, 255, 255), 3)

    # 藍色實心圓圈
    cv2.circle(img,(140, 170), 15, (255, 0, 0), -1)

    # 要画的点的坐标
    points_list = [(10, 10), (20, 20), (50, 50), (75, 75), (30, 60), (72, 27)]

    for point in points_list:
        cv2.circle(img, point, point_size, point_color, thickness)
        
      
    # 画圆，圆心为：(160, 160)，半径为：60，颜色为：point_color，实心线
    circle_mid = (50, 50)
    color = (0, 255, 0) # BGR
    cv2.circle(img, circle_mid, 10, color, 0)

    show_img(img.astype(np.uint8))
```

```python
def img_processing(img):
    # do something here
    img = draw_points(img)
    
#     cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]])

#     return img
```

```python
shape = (150, 200, 3) # y, x, RGB
origin_img = np.zeros(shape, np.uint8)
# origin_img = np.full(shape, 255).astype(np.uint8)

print("origin picture:")
show_img(origin_img)

result_img = img_processing(origin_img)
# show_img(result_img)
```
