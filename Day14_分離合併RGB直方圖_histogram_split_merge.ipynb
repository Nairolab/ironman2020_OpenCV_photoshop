```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from IPython.display import clear_output
```

```python
def show_img(img):
#     plt.figure(figsize=(15,15)) 
#     plt.figure(figsize=(10,10)) 
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.show()
```

```python
def show_histogram(img):
    # 畫出 RGB 三種顏色的分佈圖
    color = ('b','g','r')
    plt.style.use('dark_background')
    plt.figure(figsize=(10,5))
    for idx, color in enumerate(color):
        histogram = cv2.calcHist([img],[idx],None,[256],[0, 256])
        plt.plot(histogram, color = color)
        plt.xlim([0, 256])
    
    plt.show()
```

```python
def merge_RGBThreeChannel(R, G, B):
    img = cv2.merge([B, G, R])
    
    return img
```

```python
def split_RGBThreeChannel(img):
    (B, G, R) = cv2.split(img) # 3 channel
    
    # make all zeros channel
    zeros = np.zeros(img.shape[:2], dtype = np.uint8)

#     print("R channel:")
#     show_img(merge_RGBThreeChannel(R=R, G=zeros, B=zeros))
#     print("G channel:")
#     show_img(merge_RGBThreeChannel(R=zeros, G=G, B=zeros))
#     print("B channel:")
#     show_img(merge_RGBThreeChannel(R=zeros, G=zeros, B=B))

    plt.figure(figsize=(15,15)) 
    # 畫第一張圖
    plt.subplot(1, 3, 1)
    R_channel = merge_RGBThreeChannel(R=R, G=zeros, B=zeros)
    R_img = cv2.cvtColor(R_channel, cv2.COLOR_BGR2RGB)
    plt.imshow(R_img)
    plt.title("R channel image", {'fontsize':20})  

    # 畫第二張圖
    plt.subplot(1, 3, 2)
    G_channel = merge_RGBThreeChannel(R=zeros, G=G, B=zeros)
    G_img = cv2.cvtColor(G_channel, cv2.COLOR_BGR2RGB)
    plt.imshow(G_img)
    plt.title("G channel image", {'fontsize':20})
    
    # 畫第三張圖
    plt.subplot(1, 3, 3)
    B_channel = merge_RGBThreeChannel(R=zeros, G=zeros, B=B)
    B_img = cv2.cvtColor(B_channel, cv2.COLOR_BGR2RGB)
    plt.imshow(B_img)
    plt.title("B channel image", {'fontsize':20})
    
    plt.show()
    
    
    return R, G, B
```

```python
def img_processing(img):
    # do something here

    print("get RGBThreeChannel:")
    R, G, B = split_RGBThreeChannel(img)
    
    print("merge RGBThreeChannel:")
    img = merge_RGBThreeChannel(R=R, G=G, B=B)
    show_img(img)
    
    print("show histogram:")
    show_histogram(img)
    
    
    
```

```python
file_name = "./testdata/cat.jpg"
origin_img = cv2.imread(file_name)
print("origin picture:")
show_img(origin_img)

img_processing(origin_img)
```
