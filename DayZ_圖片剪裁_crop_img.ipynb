```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from IPython.display import clear_output
```

```python
def show_img(img):
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.show()
```

```python
def crop_img(img):
    
    # left, right
    x_l, x_r = 583, 983 
    
    # up, down
    y_u, y_d = 240, 640 

    # crop image
    crop_img = img[y_u:y_d, x_l:x_r]  # notice: first y, then x
    
    return crop_img
```

```python
def img_processing(img):
    # do something here
    img = crop_img(img)

    return img
```

```python
file_name = "./testdata/backkiller.jpg"
origin_img = cv2.imread(file_name)
show_img(origin_img)

result_img = img_processing(origin_img)
show_img(result_img)
```
