```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from IPython.display import clear_output
```

```python
def show_img(img):
#     plt.figure(figsize=(15,15))
    plt.style.use('dark_background')
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.show()
```

```python
def Bilateral_Filter(image, kernel_size=5):
    blurred = cv2.bilateralFilter(image, kernel_size, 75, 75)
    return blurred
```

```python
def median_Blur(image, kernel_size=5):
    blurred = cv2.medianBlur(image, kernel_size)
    return blurred
```

```python
def gaussian_Blur(image, kernel_size=5):
    blurred = cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)
    return blurred
```

```python
def averaging_Blur(image, kernel_size=5):
    blurred = cv2.blur(image, (kernel_size, kernel_size))
    return blurred
```

```python
def img_processing(img):
    # do something here
    kernel_size = 7
    
    print("Averaging Blur:")
    averagingBlur = averaging_Blur(img, kernel_size)
    show_img(averagingBlur)
    
    print("Gaussian Filtering:")
    gaussianBlur = gaussian_Blur(img, kernel_size)
    show_img(gaussianBlur)
    
    print("Median Filtering:")
    medianBlur = median_Blur(img, kernel_size)
    show_img(medianBlur)
    
    print("Bilateral Filtering:")
    BilateralFilter = Bilateral_Filter(img, kernel_size)
    show_img(BilateralFilter)

    
    # ---------------- 印出結果圖表 ---------------- #

    plt.figure(figsize=(15,10)) 
    
    
    plt.subplot(2, 2, 1)
    plt.imshow(cv2.cvtColor(averagingBlur, cv2.COLOR_BGR2RGB))
    plt.title("Averaging Blur (k = {})".format(kernel_size), {'fontsize':20})  

    
    plt.subplot(2, 2, 2)
    plt.imshow(cv2.cvtColor(gaussianBlur, cv2.COLOR_BGR2RGB))
    plt.title("Gaussian Filtering (k = {})".format(kernel_size), {'fontsize':20})
    
    
    plt.subplot(2, 2, 3)
    plt.imshow(cv2.cvtColor(medianBlur, cv2.COLOR_BGR2RGB))
    plt.title("Median Filtering (k = {})".format(kernel_size), {'fontsize':20})
    
    
    plt.subplot(2, 2, 4)
    plt.imshow(cv2.cvtColor(BilateralFilter, cv2.COLOR_BGR2RGB))
    plt.title("Bilateral Filtering (k = {})".format(kernel_size), {'fontsize':20})
    
    plt.show()
```

```python
file_name = "./testdata/cat.jpg"
origin_img = cv2.imread(file_name)
print("origin picture:")
show_img(origin_img)

img_processing(origin_img)
```
