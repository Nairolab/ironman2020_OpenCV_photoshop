```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from IPython.display import clear_output
```

```python
def show_img(img):
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.show()
```

```python
def crop_img(img):
    
    # left, right
    x_l, x_r = 583, 983 
    
    # up, down
    y_u, y_d = 240, 640 

    # crop image
    crop_img = img[y_u:y_d, x_l:x_r]  # notice: first y, then x
    
    return crop_img
```

```python
def rotate_img(img):
    (h, w, d) = img.shape
    center = (w // 2, h // 2)
    
    # 第一個參數旋轉中心，第二個參數旋轉角度(-順時針/+逆時針)，第三個參數縮放比例
    M = cv2.getRotationMatrix2D(center, 15, 1.0)
    
    # 第三個參數變化後的圖片大小
    rotate_img = cv2.warpAffine(img, M, (w, h))
    
    return rotate_img
```

```python
def resize_img(img):

    scale_percent = 50 # percent of original size
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    dim = (width, height)  
    resize_img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)  
        
    return resize_img
```

```python
def img_processing(img):
    # do something here
    img = crop_img(img)
    print("crop_img:")
    show_img(img)
    
    img = rotate_img(img)
    print("rotate_img:")
    show_img(img)
    
    img = resize_img(img)
    print("resize_img:")
    show_img(img)
    
    return img
```

```python
file_name = "./testdata/backkiller.jpg"
origin_img = cv2.imread(file_name)
print("origin_img:")
show_img(origin_img)

result_img = img_processing(origin_img)
# show_img(result_img)
cv2.imwrite('./test_result/out.jpg', result_img)
```
