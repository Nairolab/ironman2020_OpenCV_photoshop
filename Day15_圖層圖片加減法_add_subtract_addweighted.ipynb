```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
from IPython.display import clear_output
```

```python
def show_img(img):
    plt.style.use('dark_background')
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.show()
```

```python
def img_processing(img):
    # do something here

    print("Create mask:")
    # make all 128 channel
    mask = np.full(img.shape, 128).astype(np.uint8)
    show_img(mask)
    
    print("Add image with mask (img + mask):")
    add_result = cv2.add(img, mask)
    show_img(add_result)
    
    print("Subtract mask from image (img - mask):")
    subtract_result = cv2.subtract(img, mask)   
    show_img(subtract_result)
    
    print("Overlapping image and mask = 5:5")
    overlapping55 = cv2.addWeighted(img, 0.5, mask, 0.5, 0)
    show_img(overlapping55)
    
    print("Overlapping image and mask = 8:2")
    overlapping82 = cv2.addWeighted(img, 0.8, mask, 0.2, 0)
    show_img(overlapping82)
    
    # ---------------- 印出結果圖表 ---------------- #

    plt.figure(figsize=(15,10)) 
    
    
    plt.subplot(2, 2, 3)
    plt.imshow(cv2.cvtColor(add_result, cv2.COLOR_BGR2RGB))
    plt.title("Add image with mask (img + mask)", {'fontsize':20})  

    
    plt.subplot(2, 2, 4)
    plt.imshow(cv2.cvtColor(subtract_result, cv2.COLOR_BGR2RGB))
    plt.title("Subtract mask from image (img - mask)", {'fontsize':20})
    
    
    plt.subplot(2, 2, 1)
    plt.imshow(cv2.cvtColor(overlapping55, cv2.COLOR_BGR2RGB))
    plt.title("Overlapping image and mask = 5:5", {'fontsize':20})
    
    
    plt.subplot(2, 2, 2)
    plt.imshow(cv2.cvtColor(overlapping82, cv2.COLOR_BGR2RGB))
    plt.title("Overlapping image and mask = 8:2", {'fontsize':20})
    
    plt.show()
    
```

```python
file_name = "./testdata/cat.jpg"
origin_img = cv2.imread(file_name)
print("origin picture:")
show_img(origin_img)

img_processing(origin_img)
```
